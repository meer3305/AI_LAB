# Python Data Types
# Lists

```python
l1=[1,3,2,4,'WOW']
print(l1)
l1.append("hehe")
print(l1)
l1.insert(3,98)
l1[2]="HELLO"
print(l1)
```

```python
l2=l1+["lab",90,66,69]
l2.append(['AI lab',9])
print(l2)
print(len(l2))
```

```python
l2.extend(['Python',2.5])
print(l2)
```

```python
l2[5]="Artificial intelligence"
print(l2)
print(len(l2))
```

```python
del(l2[2])
print(l2)
print(len(l2))
```

```python
l1=[20,3,42,2,1,6,7,2,21]
print(l1)
l1.sort()
print(l1)
l1.reverse()
print(l1)
print(sum(l1))
```

```python
print(min(l1))
print(max(l1))
```

```python
print(type(l2))
```

```python
l3=[8,99,[2.3,6.7],['First','Lab']]
print(l3[2][0])
print(l3[2][1])
print(l3[3][0])
print(l3[3][1])
```

# Tuples

```python
t1=(1,5,22,4,'SHEH')
print(t1)
print(t1[-1])
```

```python
t2=t1+(9,3,2,4)
print(t2)
```

```python
#Slicing
t2=(10,'AI','LAB',2.85,56.3)
print(t2[0:3])
print(t2[3:5])
```

# Dictionary

```python
dict={"brand":"Ford","Model":"Mustang","Year":1964}
print(dict)
```

```python
dict["Australia"]=300
print(dict)
print(len(dict))
```

```python
print(dict.keys())
print(dict.values())
```

# SETS

```python
s1={2,4,1,3,25}
print(s1)
s1.add(21)
print(s1)
```

```python
s2={5,8,4,1,9,0}
print(s2)
s2.add(67)
print(s2)
```

```python
s3=s1&s2
s4=s1|s2
s5=s1-s2
print(s3)
print(s4)
print(s5)
```
