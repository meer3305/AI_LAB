# Breadh First Search

# #Undirected Graph

```python
graph = {
    '5': ['3', '7'],
    '3': ['2', '4', '5'],
    '7': ['5', '8'],
    '2': ['3'],
    '4': ['3', '8'],
    '8': ['4', '7']
}
visit=[]
queue=[]

def bfs(visit, graph, node):
    visit.append(node)
    queue.append(node)

    while queue:
        m = queue.pop(0)
        print(m, end=" ")

        for neighbour in graph[m]:
            if neighbour not in visit:
                visit.append(neighbour)
                queue.append(neighbour)

print("Following is Breadth First Search:")
visited = []
bfs(visited, graph, '3')
```

# Directed Graph

```python
graph = {
    '5': ['3', '7'],
    '3': ['2', '4',],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}


def bfsdirected(graph, initial):
    visit=[]
    queue=[initial]

    while queue:
        node = queue.pop(0)
        if node not in visit:
            visit.append(node)
            neighbours=graph[node]
            for neighbour in neighbours:
                queue.append(neighbour)
    return visit
            

print("Following is Breadth First Search:")
bfsdirected(graph, '3')
```

# Depth First Search

# directed Graph

```python
graph = {
    '5': ['3', '7'],
    '3': ['2', '4',],
    '7': ['8'],
    '2': [],
    '4': ['8'],
    '8': []
}
visited=set()
def dfs(visited,graph,node):
    if node not in visited:
        print(node,end=" ")
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited,graph,neighbour)
print("Following is the Depth First Search")
dfs(visited,graph,'3')
```
